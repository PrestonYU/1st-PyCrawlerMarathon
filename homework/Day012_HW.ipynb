## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
```

```python
print(r)
```

```python
soup
```

```python
target = soup.find(class_="part_list_2").find_all('h3')
```

### ① 取出今天所有的發文

```python
# 今日日期
pattern = re.compile('2020/01/19')
```

```python
for d in target:
    if re.match(pattern, d.find(class_="date").text):
        pass
    else:
        continue
    print(d.find(class_="date").text,d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
datelist = []
typelist = []
title = []
```

```python
for d in target:
    if re.match(pattern, d.find(class_="date").text):
        pass
    else:
        continue
    datelist.append(d.find(class_="date").text)
    typelist.append(d.find_all('em')[-1].text)
    title.append(d.find_all('a')[-1].text)
    print(d.find(class_="date").text,d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```

### ③ 哪一個類別的文章最多

```python
df = pd.DataFrame({'date': datelist,'type': typelist,'title': title})
df
```

```python
a = df.groupby('type').size()
print(f'最多的類別為{a.idxmax()},數量為:{max(a)}')
```
