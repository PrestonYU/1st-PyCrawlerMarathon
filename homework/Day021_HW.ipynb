## 1. 取出今天所有的新聞

```python
# 打開瀏覽器
import re
import time
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
browser = webdriver.Chrome(executable_path='/Applications/Google Chrome.app/Contents/MacOS/chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 選取今天時間
Day = Select(browser.find_element_by_id('selD'))
Day.select_by_value('21')
browser.find_element_by_id('button').click()
```

```python
# 利用 selenium 持續執行下滑的動作
SCROLL_PAUSE_TIME = 0.5
last_height = browser.execute_script("return document.body.scrollHeight")
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # load page
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
pattern = re.compile('2020/01/21')
```

```python
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")
for d in soup.find(class_="part_list_2").find_all('h3'):
    if re.match(pattern, d.find(class_="date").text):
        pass
        print(d.find(class_="date").text,d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime, timedelta

date_time = datetime.now() - timedelta(0)
date_time = date_time.strftime("%Y/%m/%d %H:%M:%S")
print("取得現在時間：", date_time)
```

```python
current_time_hour = datetime.now().strftime("%H")
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = time.strptime(d.find(class_='date').text ,'%Y/%m/%d %H:%M')
    if (date.tm_mday == 21) & ((date.tm_hour == int(current_time_hour)-2) or (date.tm_hour == int(current_time_hour)-1) or (date.tm_hour == int(current_time_hour))):
        print(d.find(class_="date").text,d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```
